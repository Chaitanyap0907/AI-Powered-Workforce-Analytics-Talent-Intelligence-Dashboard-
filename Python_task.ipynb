```python
import pandas as pd
import numpy as np

# Set a seed for reproducibility
np.random.seed(42)

# Generate synthetic data for 10 rows and 10 columns
data = {
    'transaction_id': [f'TR{i:03d}' for i in range(1, 11)],
    'customer_id': [f'CUST{np.random.randint(100, 200)}' for _ in range(10)],
    'product_id': [f'PROD{np.random.randint(1000, 2000)}' for _ in range(10)],
    'quantity': np.random.randint(1, 5, 10),
    'price': np.round(np.random.uniform(5.0, 100.0, 10), 2),
    'transaction_date': pd.to_datetime(pd.date_range(start='2023-01-01', periods=10, freq='D')),
    'payment_method': np.random.choice(['Credit Card', 'PayPal', 'Debit Card', 'Bank Transfer'], 10),
    'shipping_address': [f'Address {i}' for i in range(1, 11)],
    'order_status': np.random.choice(['Completed', 'Pending', 'Shipped'], 10),
    'category': np.random.choice(['Electronics', 'Apparel', 'Books', 'Home & Kitchen', 'Beauty'], 10)
}

df = pd.DataFrame(data)

# Display the DataFrame
display(df)
```
